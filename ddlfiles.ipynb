```python
import requests
from bs4 import BeautifulSoup

HLTV_TOURNAMENT_URL = "https://www.hltv.org/events/8034/iem-katowice-2025"  
# Replace with correct event ID
BASE_URL = "https://www.hltv.org"

def get_match_links():
    headers = {"User-Agent": "Mozilla/5.0"}
    response = requests.get(HLTV_TOURNAMENT_URL, headers=headers)
    soup = BeautifulSoup(response.text, "html.parser")
    print(soup.prettify())

    match_links = []
    for match in soup.find_all("a", class_="a-reset"):
        link = match.get("href")
        if link and "/matches/" in link:
            match_links.append(BASE_URL + link)

    return match_links

match_links = get_match_links()
print(f"Found {len(match_links)} matches.")
```
